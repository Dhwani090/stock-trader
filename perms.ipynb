```python
watchlists = {
    '8/14': [
        "NVMI", "DUOL", "CVNA", "SE", "FOUR", "NTRA",
        "AZPN", "IESC", "SHOP", "CAVA", "MNDY", "BMA",
        "KRUS", "PAM", "RARE", "LRN", "MASI", "ICUI",
        "FTNT", "AXON"
    ],
    '8/15': [
        "NVMI", "APP", "LLY", "Z", "TAYD", "SHOP", "NTRA", "DUOL",
        "IESC", "NRG", "SE", "MNDY", "CAVA", "FOUR", "BMA",
        "ICUI", "PAM", "MASI", "LRN", "RARE"
    ],
    '8/16': [
        "NVMI", "APP", "LLY", "Z", "TAYD", "SHOP", "NTRA", "DUOL", "IESC", "NRG",
        "SE", "MNDY", "CAVA", "FOUR", "BMA", "ICUI", "PAM", "MASI", "LRN", "RARE"
    ]
}


# Define the ranges for rsi_window and buy_rsi_threshold
rsi_window_range = range(3, 16)  # Example range from 5 to 10
buy_rsi_threshold_range = range(30, 81, 5)  # Example range from 30 to 70 in steps of 5


# Dates and their corresponding watchlists
dates_watchlists = [
    {
        'date': '2024-08-14',
        'start_time': datetime(2024, 8, 14, 9, 50),
        'watchlist': watchlists['8/14'],
    },
    {
        'date': '2024-08-15',
        'start_time': datetime(2024, 8, 15, 9, 50),
        'watchlist': watchlists['8/15'],
    },
    {
        'date': '2024-08-16',
        'start_time': datetime(2024, 8, 16, 9, 50),
        'watchlist': watchlists['8/16'],
    }
]


# Initialize variables to track the best configuration
best_config = None
highest_profit = float('-inf')


# Iterate through all permutations of rsi_window and buy_rsi_threshold
for rsi_window, buy_rsi_threshold in product(rsi_window_range, buy_rsi_threshold_range):
    # Initialize the cumulative profit for this configuration
    cumulative_profit = 0


    # Test this configuration across all specified dates
    for date_config in dates_watchlists:
        # Set up the config in FRtest
        FRTEST.CONFIG = {
            'rsi_window': rsi_window,
            'stoch_rsi_window': 15,
            'buy_rsi_threshold': buy_rsi_threshold,
            'capital': 25000,
            'target_percentage': 0.55,  
            'stop_loss_percentage': 0.5,
            'stage2tp': 1,
            'stage2sl': 0.5,
            'stage2': False,
            'activateS2': False,
            'reorder': True,
            'Trail': False,
            'folder_name': 'LiveDataTest',
            'watchlist': date_config['watchlist'],
            'start_time': date_config['start_time'],
            'test_date': date_config['date'],
            'clicks': 370,
            'lower-limit': -375,
            'trailing-loss': 200,
            'trailing-breakthrough': 300,
            'trailing': False,
            'trading': True,
            'TSL': 0,
            'TSL_D': 0.55
        }


        # Run the full program
        FRTEST.main()  # Replace bot() with main() or equivalent function


        # After the bot runs, get the profit from data_stats
        profit = FRTEST.data_stats['Profit']
        #profit = 0
        cumulative_profit += profit


    # Check if this configuration yields the highest cumulative profit
    if cumulative_profit > highest_profit:
        highest_profit = cumulative_profit
        best_config = {
            'rsi_window': rsi_window,
            'buy_rsi_threshold': buy_rsi_threshold,
            'cumulative_profit': cumulative_profit
        }


    print(f"Tested config: RSI Window={rsi_window}, Buy RSI Threshold={buy_rsi_threshold}, Cumulative Profit={cumulative_profit}")


# Output the best configuration and its profit
print(f"\nBest configuration:")
print(f"RSI Window: {best_config['rsi_window']}")
print(f"Buy RSI Threshold: {best_config['buy_rsi_threshold']}")
print(f"Highest Cumulative Profit: {best_config['cumulative_profit']}")
```
